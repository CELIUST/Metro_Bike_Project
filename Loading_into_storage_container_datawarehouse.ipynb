#Loading into storage container in the datawarehouse


```python
# Azure Storage Account information

account_name = "Censored"
account_key = "Censored"
container_name = "metrobikedata"
local_file_path = '/Users/tam/Downloads/Metro_bike_data_clean'
blob_name = "Metro_bike_data_clean.csv"

# Create a BlobServiceClient
blob_service_client = BlobServiceClient(account_url=f"https://{account_name}.blob.core.windows.net", credential=account_key)

# Get a reference to the container
container_client = blob_service_client.get_container_client(container_name)

# Specify the local directory path
local_directory_path = '/Users/tam/Downloads/Metro_bike_data_clean'

# Iterate through files in the directory
for filename in os.listdir(local_directory_path):
    file_path = os.path.join(local_directory_path, filename)

    # Check if the item is a file (not a subdirectory)
    if os.path.isfile(file_path):
        # Open and upload the file to Azure Data Lake Storage
        with open(file_path, "rb") as data:
            container_client.upload_blob(name=filename, data=data)

            print(f"File '{filename}' uploaded to Azure Data Lake Storage.")
```
